# Replace one data with new data

```python
# replace one data with new data

mongo_collection.update_one({'companyName': 'iNeuron'}, {"$set":{'companyName': 'pwskill'}})
```

```python
for i in mongo_collection.find({'companyName': 'pwskill'}):
    print(i)
```

```python
# replace many  data with new data

mongo_collection.update_many({'companyName': 'iNeuron'}, {"$set":{'companyName': 'pwskill'}})
```

# Condition

```python
#condition finding the id which is greater than or equal 4

for i in mongo_collection.find({"_id" :{"$gte" : '4'}}):
    print(i)
```

```python
#condition finding the id which is lower than 5
for i in mongo_collection.find({"_id":{"$lt":"5"}}):
    print(i)
```
